### Plot Number of Unique Genes Expressed Per Cell

```python
# Create plot
fig, axes = plt.subplots(nrows = 2, ncols = 2, figsize = (18,13))

# Filter out cells with fewer than this many genes expressed
gene_filter = 500

# Pre-filter
n_cells_pre = len(Adata)
for ax in [axes[0,0], axes[1,0]]:
    ax.scatter(range(len(Adata)), Adata.obs['nGenes'])              # x = cell number, y = Number of Unique genes
    ax.axhline(gene_filter, color="red")                            # Keep cells with at least this many unique genes
    ax.set_title('Number of Unique Genes per Cell (Pre-filter)')
    
# Post-filter
Adata2 = Adata[Adata.obs['nGenes'] > gene_filter, :]                 # Filter the cells
for ax in [axes[0,1], axes[1,1]]:
    ax.scatter(range(len(Adata2)), Adata2.obs['nGenes'])
    ax.axhline(gene_filter, color = "red")                      
    ax.set_title('Number of Unique Genes per Cell (Post-filter)')
    
# Define partitions to separate by sample for second graph
lengths = [0] * (len(samples)-1)
for i,sample in enumerate(samples[:-1]):
    lengths[i] = np.sum(Adata2.obs['sample_num'] == i+1)
    
lengths = np.cumsum(lengths)

for l in lengths:
    axes[1,0].axvline(l, color = 'black')
    axes[1,1].axvline(l, color = 'black')
    
    
    
# Set axes labels and titles
for ax in axes.flatten():
    ax.set_xlabel('Cell Number')
    ax.set_ylabel('Number of Genes')
    ax.set_ylim(0, 7000)
    
# Statistics
n_cells_post = len(Adata2)
diff = n_cells_pre - n_cells_post
percent = (diff)/n_cells_pre * 100

text1 = 'There were {} cells before the filter.'.format(n_cells_pre)
text2 = 'There were {} cells with fewer than {} unique genes expressed.'.format(diff, gene_filter)
text3 = 'There are {} cells remaining, so {:.2f}% of cells were lost.'.format(n_cells_post,percent)
text = '\n'.join([text1, text2, text3])

print(text)
```

# Apply housekeeping genes filter

```python
# Housekeeping genes filter

# Filter works by creating a new Adata4.obs called 'housekeeping'.
# 'housekeeping' is a single value for each cell that calculates the percentage of housekeeping genes (given
# from a list) that appears in a cell. If the gene is an expression count greater than 0, then the
# gene is considered to be expressed.

# For each cell, if greater than 50% of housekeeping are expressed
# in the cell, then we keep the cell. Otherwise, we filter it out.

# Start making plots
fig, axes = plt.subplots(ncols = 2, figsize = (18,8))

n_cells_pre = len(Adata4)     # Number of cells before filtering




# Start filtering process, define filter percentage
hskp_filter = 0.50

# Import housekeeping gene list
hskp_genes = open('data/GeneLists/HousekeepingGenes.txt','r')
hskp_genes = hskp_genes.read().split(",")

# Keep intersection of housekeeping genes and genes that appear in the Adata4 structure
hskp_genes = IntersectionOfGenes(Adata4, hskp_genes)

# Find the percentage of housekeeping genes expressed by each cell
percent_hskp = np.count_nonzero(Adata4[:, hskp_genes].X, axis=1) / len(hskp_genes)

# Make first plot
axes[0].scatter(range(len(Adata4)), percent_hskp)
axes[0].set_title('Percent of Housekeeping Genes per Cell (Pre-filter)')

# Extract Adata4 cells
Adata5 = Adata4[percent_hskp > hskp_filter, :]
percent_hskp = np.count_nonzero(Adata5[:, hskp_genes].X, axis=1) / len(hskp_genes)

# Make second plot
axes[1].scatter(range(len(Adata5)), percent_hskp)
axes[1].set_title('Percent of Housekeeping Genes per Cell (Post-filter)')

# Labels
for ax in axes:
    ax.set_xlabel('Cell Number')
    ax.set_ylabel('Percent of Housekeeping Genes')
    ax.axhline(hskp_filter, color = "red")
    ax.set_ylim(0, 1)

# Statistics
n_cells_post = len(Adata5)
diff = n_cells_pre - n_cells_post
percent = (diff)/n_cells_pre * 100

text1 = 'There were {} cells before the filter.'.format(n_cells_pre)
text2 = 'There were {} cells that expressed less than {} percent of housekeeping genes.'.format(diff, hskp_filter*100)
text3 = 'There are {} cells remaining, so {:.2f}% of cells were lost.'.format(n_cells_post,percent)
text = '\n'.join([text1, text2, text3])

print(text)
```

# Recount Clone Sizes

```python
# Count how many times each TCR is found in the data

# For all AB cells, add the mouse name to the TCR
# For all AB cells, rename the TCR to TCR_mouse
Adata.obs['TCR'] = Adata.obs['TCR'].astype(str)
Adata.obs.loc[Adata.obs['AB_status'] == 'AB_cell', 'TCR'] = \
Adata.obs.loc[Adata.obs['AB_status'] == 'AB_cell', 'TCR'].astype(str) + '_' + \
Adata.obs.loc[Adata.obs['AB_status'] == 'AB_cell', 'sample'].astype(str)
Adata.obs['TCR'] = Adata.obs['TCR'].astype('category')

Adata.obs['CloneSize'] = Adata.obs.groupby('TCR')['TCR'].transform('count')

# Set non-AB cells to clone size of 0
Adata.obs.loc[ Adata.obs['AB_status'] != 'AB_cell', 'CloneSize' ] = 0
```

# Create AnnData structure, adata_hv, of only highly variable genes

```python
# Make a copy to run through the next few steps
adata_hv = Adata.copy()
```

# Normalize and Logarithmize Data

```python
# Total-count normalize to 10,000 reads per cell
sc.pp.normalize_total(adata_hv, target_sum = 1e4)

# Logarithmize the data
sc.pp.log1p(adata_hv)

print(adata_hv)
```

```python
# Identify highly variable genes
sc.pp.highly_variable_genes(adata_hv, min_mean=0.0125, max_mean=3, min_disp=0.5)

# Plot
sc.pl.highly_variable_genes(adata_hv)
```

```python
# Filter by keeping only highly variable genes
n_genes_pre = len(adata_hv.var_names)
adata_hv = adata_hv[:, adata_hv.var['highly_variable']]
n_genes_post = len(adata_hv.var_names)
n_genes_filtered = n_genes_pre - n_genes_post
percent_genes_filtered = n_genes_filtered / n_genes_pre * 100


print("Before the filter, there were %d genes total." % n_genes_pre )
print("There were %d genes that were removed because they were not highly variable." % n_genes_filtered)
print("%.2f%% of genes were removed from the data." % percent_genes_filtered)
print("After the filter, there are now %d genes total." % n_genes_post )
print("There are still %d cells total." % len(list(adata_hv.obs_names)))
```

# Principal component analysis for adata (highly variable genes)

```python
# Remove dimensionality of data, find main axis of variation, and denoise data with PCA
sc.tl.pca(adata_hv, svd_solver='arpack')

# Find how many PCs contribute to the total variance and display this on a scree plot
sc.pl.pca_variance_ratio(adata_hv, log=True)
```

# Create kNN and Leiden clusters for adata (highly variable genes)

```python
# Define a color palette given
color_palette = ['red','black','blue','green','orange','yellow','brown','magenta','pink','gray','purple',
                 'tan','lightblue','lightgreen','maroon','aqua','lime','yellow']
```

```python
# Compute neighborhood graph
sc.pp.neighbors(adata_hv, n_neighbors=10, n_pcs=40)

# UMAP graphs
sc.tl.umap(adata_hv)

# Leiden cluster
sc.tl.leiden(adata_hv, resolution = 0.5)

sc.pl.umap(adata_hv, color = ['leiden'], palette = color_palette)
```
